```python
%matplotlib notebook
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
from pylab import *


hdul=fits.open('L559141_SAP000_natural_1500l.fits')

#with fits.open('L559141_SAP000_natural_1500l.fits') as hdul:
#    hdul.info()
#list(hdr.keys())

hdul.info()
hdr = hdul[0].header
#hdul[0].header[0]
image=fits.getdata('L559141_SAP000_natural_1500l.fits')

print('Dimension of the file:', image.shape)

fig1=figure(1)
plt.imshow(image[0][0], cmap='hot')
plt.title('Stokes I of first frame')
plt.colorbar()
fig1.savefig('stokes1_frame1.png')
```

```python
# maximum intensity at x=254, y=153
#spectrum of this pixel

fig2=figure(2)
plt.plot(image[0, :, 153,254])
plt.title('Spectrum of brightest pixel in Stokes I')
plt.xlabel('Frequency Channel')
plt.ylabel('Flux-Density')
fig2.savefig('spectra_254_153.png')
```

```python
# Stokes V frames
fig3=figure(3)
plt.imshow(image[1][0], cmap='hot')
plt.title('Stokes V of first frame')
plt.colorbar()
fig3.savefig('stokes1_frame1.png')
```
